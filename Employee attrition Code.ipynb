# Label encoding

```python
# encode Target variable(Attrition)
df['Attrition']=df['Attrition'].replace({'No':0,'Yes':1})
```

```python
#encode binary variables
df['OverTime']=df['OverTime'].map({'No':0,'Yes':1})
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
```

```python
df['Over18']=df['Over18'].map({'Y':1,'N':0})
```

```python
#encode categorical columns which are ordinal(create dummy variables)
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders={}
for column in encoding_cols:
    label_encoders[column]=LabelEncoder()
    df[column]=label_encoders[column].fit_transform(df[column])
df.head()
    
```

```python
#attrition %
num_left = df['Attrition'].sum()
total_employees = len(df)
attrition_rate = (num_left / total_employees) * 100
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
from scipy.stats import ttest_ind, chi2_contingency

#Hypothesis testing
attrition_group = df[df['Attrition'] == 1]
non_attrition_group = df[df['Attrition'] == 0]

#Two-Sample T-Test
numerical_cols = ['Age', 'DistanceFromHome', 'MonthlyIncome', 'NumCompaniesWorked', 
                  'PercentSalaryHike', 'TotalWorkingYears', 'YearsAtCompany', 
                  'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']

print("T-Tests between Attrition and Non-Attrition Groups for Numerical Variables:\n")
for col in numerical_cols:
    t_stat, p_val = ttest_ind(attrition_group[col], non_attrition_group[col])
    print(f"{col}: t-statistic = {t_stat:.4f}, p-value = {p_val:.4f}")

#Chi-Square Test
categorical_cols = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 
                    'JobRole', 'MaritalStatus', 'OverTime']

print("\nChi-Square Test Results for Categorical Variables:\n")
for col in categorical_cols:
    contingency_table = pd.crosstab(df[col], df['Attrition'])
    chi2, p_val, dof, expected = chi2_contingency(contingency_table)
    print(f"{col}: chi2 = {chi2:.4f}, p-value = {p_val:.4f}")
```

```python
#Correlation Matrix for numerical features
correlation_matrix = df[numerical_cols + ['Attrition']].corr()

#Heatmap for Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix with Attrition")
plt.show()

print("\nCorrelation of Numerical Features with Attrition:\n")
correlation_attrition = correlation_matrix['Attrition'].drop('Attrition')
print(correlation_attrition.sort_values(ascending=False))
```

# Machine Learning

```python
selected_features = ['Age','DistanceFromHome','YearsAtCompany','YearsWithCurrManager','YearsInCurrentRole','JobRole',
                     'MaritalStatus','OverTime']
```

```python
X = df[selected_features] #Input features
y=df['Attrition'].values#Output
```

# Resampling

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus=RandomOverSampler(random_state=42)
X_over,y_over=rus.fit_resample(X,y)
print(Counter(y_over))
```

```python
#Split data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test=train_test_split(X_over,y_over,test_size=0.2,random_state=42)
```

```python
print(X_train.shape) 
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
prediction=logreg.predict(X_test)
cnf_matrix=confusion_matrix(y_test,prediction)
print("Accuracy score:",accuracy_score(y_test,prediction))
```

```python
#heatmap
fig=plt.figure(figsize=(15,6))
ax1=fig.add_subplot(1,2,1)
ax1=sns.heatmap(pd.DataFrame(cnf_matrix),annot=True,cmap='Blues',fmt='d')
bottom,top=ax1.get_ylim()
ax1.set_ylim(bottom+0.5,top-0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

#roc curve
ax2=fig.add_subplot(1,2,2)
y_pred_proba=logreg.predict_proba(X_test)[::,1]
fpr,tpr,_=roc_curve(y_test,prediction)
auc=roc_auc_score(y_test,prediction)

ax2.plot(fpr, tpr, label='data 1, auc=' + str(auc))
ax2.plot([0, 1], [0, 1], 'k--')  # Diagonal line for ROC
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc=4)

plt.tight_layout()
plt.show()
```
